```python
dataset_list = ["conc", "ener", "nava", "powe", "prot", "yach"]
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(6, 3, sharex=True)
for idx_d, dataset in enumerate(dataset_list):
    data = UCIDataset(dataset)
    in_dim = data.Xtrain.shape[1]
    out_dim = data.Ytrain.shape[1]
    models = [MLP(in_dim, [25, 25], out_dim), MLP_bn(in_dim, [25, 25], out_dim), MLP_drop(in_dim, [25, 25], out_dim)]
    for idx_m, model in enumerate(models):
        loss = train_nn(model, data, 1000)
        axs[idx_d][idx_m].plot(loss)
```
